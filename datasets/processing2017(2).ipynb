## As we have to merge new_metaData and castAndCrew, we must have same datatypes

```python
new_metaData = mov_metaData.loc[mov_metaData.year == 2017,['genres','id','title','year']].astype(str)
```

```python
castAndCrew = castAndCrew.astype(str)
```

```python
new_metaData.head()
```

```python
castAndCrew['id'].dtypes
```

```python
new_metaData['id'].dtypes
```

```python
data = pd.merge(new_metaData, castAndCrew, on='id')
```

```python
data.head()
```

```python
print(data['genres'].dtypes, new_metaData['genres'].dtypes)
```

```python
# converting into list
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def genresHelper(x):
    genres = []
    string = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            genres.append('Sci-Fi') # trying to short the name
        else:
            genres.append(i.get('name'))
    if genres == []:
        return np.NaN
    else:
        return (string.join(genres))
```

```python
data['genres_list'] = data['genres'].map(lambda x: genresHelper(x))
```

```python
pd.DataFrame(data['genres_list']).head()
```

```python
def actor1Helper(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: actor1Helper(x))
```

```python
def actor2Helper(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: actor2Helper(x))
```

```python
def actor3Helper(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: actor3Helper(x))
```

```python
print(data['actor_1_name'][:5], data['actor_2_name'][:5], data['actor_3_name'][:5])
```

```python
def directorsHelper(x):
    directors = []
    string = " "
    for i in x:
        if i.get('job') == 'Director':
            directors.append(i.get('name'))
    if directors == []:
        return np.NaN
    else:
        return (string.join(directors))
```

```python
data['director_name'] = data['crew'].map(lambda x: directorsHelper(x))
```

```python
data['director_name']
```

```python
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
```
